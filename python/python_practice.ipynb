## Order Of Operations
The acronym PEMDAS is useful to remember the rules




1.   > Parentheses have the highest precedence and can be used to force an expression to
evaluate in the order you want. Since expressions in parentheses are evaluated first,
2 * (3-1) is 4, and (1+1)**(5-2) is 8. You can also use parentheses to make an
expression easier to read, as in (minute * 100) / 60, even if it doesn’t change the
result.
2.   > Exponentiation has the next highest precedence, so 1 + 2**3 is 9, not 27, and 2 *3**2 is 18, not 36.

3.  > Multiplication and Division have higher precedence than Addition and Subtraction.
So 2*3-1 is 5, not 4, and 6+4/2 is 8, not 5.

4. > Operators with the same precedence are evaluated from left to right (except exponen-
tiation). So in the expression degrees / 2 * pi, the division happens first and the
result is multiplied by pi. To divide by 2π, you can use parentheses or write degrees
/ 2 / pi.

```python
def sum_digits(s) -> int:
    """Sums all the number characters in the string

    Args:
        s: string
    """
    out = 0

    for char in s:
        if char.isdigit():
            out += int(char)
    return out

print(sum_digits("001 hello 3"))
```

```python
def is_anagram(w1, w2):
    w1_sorted = ''.join(sorted(w1))
    w2_sorted = ''.join(sorted(w2))
    return w1_sorted == w2_sorted

print(is_anagram('stop', 'tops'))
print(is_anagram('hello', 'world'))

def all_anagrams(word, word_list):
    word_sorted = ''.join(sorted(word))

    return [w for w in word_list if ''.join(sorted(w)) == word_sorted]

word_list = ['stop', 'pots', 'test', 'tops', 'trop', 'wall', 'malayalam', 'noon', 'rotator']
print(all_anagrams(word='stop', word_list=word_list))
```

```python
def anagrams(word_list):
    output = {}

    for word in word_list:
        root_word = ''.join(sorted(word))
        output[root_word] = output.get(root_word, []) + [word]

    return list(output.values())
words = ['deltas', 'desalt', 'lasted', 'salted', 'slated', 'staled', 'retainers',
         'ternaries','generating', 'greatening','resmelts', 'smelters', 'termless']
anagrams(words)
```

```python
def word_distance(w1, w2):
    distance = 0
    for c1, c2 in zip(w1, w2):
        if not c1 == c2:
            distance += 1
    return distance

# word_distance('hello', 'world')
print(word_distance('hello0uu', 'world0uu'))
```

```python
def metathesis_pairs(word_list):
    output = []
    _anagrams = anagrams(word_list)
    for anagram in _anagrams:
        _len = len(anagram)
        ana_metas = []
        for i in range(_len):
            for j in range(i+1, _len):
                if word_distance(anagram[i], anagram[j]) == 2:
                    meta_pairs = (anagram[i], anagram[j])
                    ana_metas.append(meta_pairs)
        if len(ana_metas) > 0:
            output.append(ana_metas)
    return output
words = ['deltas', 'desalt', 'lasted', 'salted', 'slated', 'staled', 'retainers',\
         'ternaries','generating', 'greatening','resmelts', 'smelters', 'termless']
print(metathesis_pairs(words))
```

```python
class Time:
    """Represent a time of day"""
    def __init__(self, hour, minute, second):
        self.hour = hour
        self.minute = minute
        self.second = second

    def time_to_seconds(self):
        return self.hour * 3600 + self.minute * 60 + self.second

    @staticmethod
    def is_after(t1, t2):
        return t1.time_to_seconds() > t2.time_to_seconds()

    @staticmethod
    def seconds_to_time(seconds):
        hours, seconds = divmod(seconds, 3600)
        minutes, seconds = divmod(seconds, 60)
        return Time(hours, minutes, seconds)

    @staticmethod
    def subtract_time(t1, t2):
        t1_sec, t2_sec = t1.time_to_seconds(), t2.time_to_seconds()
        if t1_sec > t2_sec:
            return Time.seconds_to_time(t1_sec - t2_sec)
        else:
            return Time.seconds_to_time(t2_sec - t1_sec)

    def __str__(self):
        return f"{self.hour}:{self.minute:02d}:{self.second:02d}"

t = Time(11,30,0)
print(t)
print(t.time_to_seconds())
print("Seconds to time: ", Time.seconds_to_time(3662))
# Objects are mutable
t.hour += 1
print(t)
```

```python
class Date:
    """Represents a year, month, and day"""
    month_mapper = {"January": 1, "February": 2, "March": 3, "April": 4, \
                    "May": 5, "June": 6, "July": 7, "August": 8, "September": 9,\
                    "October": 10, "November": 11, "December": 12}

    def __init__(self, date: int, month: str, year: int):
        assert date > 0 and date < 32, "Date invalid"
        self.date = date
        try:
            self.month = Date.month_mapper[month.capitalize()]
        except KeyError:
            raise KeyError('Invalid month: ', month)
        assert year > 0, "Invalid Year"
        self.year = year

    def __str__(self):
        return f"{self.year}-{self.month:02d}-{self.date:02d}"

    @staticmethod
    def is_after(d1, d2):
        return (d1.year, d1.month, d1.date) > (d2.year, d2.month, d2.date)
```

```python
class Card:
    """Represents a stnadard playing card."""

    suit_names = ["Clubs", "Diamonds", "Hearts", "Spades"]
    rank_names = [None, "Ace",  2, 3, 4, 5, 6, 7, 8, 9, 10, "Jack", "Queen", "King", "Ace"]

    def __init__(self, suit: int, rank: int):
        """Card object is represented by two numbers for suit and rank"""
        self.suit = suit
        self.rank = rank

    def __str__(self):
        rank_name = Card.rank_names[self.rank]
        suit_name = Card.suit_names[self.suit]
        return f"{rank_name} of {suit_name}"

    def __eq__(self, other):
        return self.suit == other.suit and self.rank == other.rank

    def __lt__(self, other):
        return self.to_tuple() < other.to_tuple()

    def to_tuple(self):
        return (self.suit, self.rank)
```

```python
import random
class Deck:
    def __init__(self, cards=[]):
        self.cards = cards

    def make_cards():
        cards = []
        for suit in range(4):
            for rank in range(2, 15):
                card = Card(suit, rank)
                cards.append(card)
        return cards

    def take_card(self):
        card = self.cards.pop()
        return card

    def put_card(self, card):
        self.cards.append(card)

    def move_cards(self, other, quantity):
        # Polymorphic function to move cards from self to other (deck/hand)
        for i in range(quantity):
            card = self.take_card()
            other.put_card(card)

    def shuffle(self):
        random.shuffle(self.cards)

    def sort(self):
        self.cards.sort()

    def __str__(self):
        res = []
        for card in self.cards:
            res.append(str(card))
        return "\n".join(res)
```

```python
class Hand(Deck):
    """Represents a hand of playing cards."""
    def __init__(self, label=''):
        self.label = label
        self.cards = []
```

```python
class BridgeHand(Hand):
    """Represents a bridge hand."""

    hcp_dict = {
        'Ace': 4,
        'King': 3,
        'Queen': 2,
        'Jack': 1,
    }

    def high_card_point_count(self):
        count = 0
        for card in self.cards:
            rank_name = Card.rank_names[card.rank]
            count += BridgeHand.hcp_dict.get(rank_name, 0)
        return count
```

```python
class PokerHand(Hand):
    """Represents a poker hand."""

    def suits_count(self):
        count = {}
        for card in self.cards:
            suit = card.suit
            count[suit] = count.get(suit, 0) + 1
        return count

    def ranks_count(self):
        count = {}
        for card in self.cards:
            rank = card.rank
            count[rank] = count.get(rank, 0) + 1
        return count

    def has_flush(self):
        _suits_count = self.suits_count()
        return any([_suits_count[key] >=5 for key in _suits_count])

    def possible_straights(self):
        _possible_straights = []
        ranks = Card.rank_names
        for i in range(1, len(ranks)-4):
            _possible_straights.append(ranks[i:i+5])
        return _possible_straights

    def has_straight(self):
        _possible_straights = self.possible_straights()
        ranks = [card.rank for card in self.cards]
        ranks.sort()
        ranks = list(set(ranks))
        for i in range(0, len(ranks)-4):
            if ranks[i:i+5] in _possible_straights:
                return True
        return False

    def has_straight_flush(self):
        if not self.has_flush():
            return False
        # Check for straights
        _possible_straights = self.possible_straights()
        # Find the flush suits
        _suits_count = self.suits_count()
        flush_suits = [suit for suit in _suits_count if _suits_count[suit] >=5]
        # Loop into all the ranks of same flush suit and check if straight
        for suit in flush_suits:
            ranks = [card.rank for card in self.cards if card.suit == suit]
            ranks.sort()
            for i in range(0, len(ranks)-4):
                if ranks[i:i+5] in _possible_straights:
                    return True
        return False

```

```python
def my_decorator(func):
    def wrapper():
        print("Before calling the function.")
        func()
        print("After calling the function.")
    return wrapper

@my_decorator
def say_hello():
    print(f"Hello there")

# say_hello()

# Decorators with argument
def my_decorator_with_arg(func):
    def wrapper(*args, **kwargs):
        print("Before calling the function.")
        func(*args, **kwargs)
    return wrapper

@my_decorator_with_arg
def say_hello_with_name(*args):
    names = " ".join(args)
    print(f"Hello {names}!")

# say_hello_with_name("Vikas", "Srivastava")

@my_decorator_with_arg
def say_hello_with_name(**kwargs):
    name = " ".join(kwargs.values())
    print(f"hello {name}")

# print(say_hello_with_name(first_name="Vikas", last_name="Srivastava"))

def log_arguments(func):
    def wrapper(*args, **kwargs):
        print("KWARGS:", kwargs)
        print(f"Arguments were: {args}, {kwargs}")
        return func(*args, **kwargs)
    return wrapper

@log_arguments
def add(*args, **kwargs):
    print("kwargs type", type(kwargs))
    print("args type: ", type(args))
    return sum(args) + sum(kwargs.values())

print(add(2, 4, num=1, ber=3))
```

Q: What is an iterator in Python?
👉 An object that lets you access elements one at a time. It implements __iter__() and __next__().

Q: How do you create an iterator?
👉 Use iter() on an iterable (like a list), or define a class with __iter__ and __next__.

Q: What happens when an iterator has no more items?
👉 StopIteration is raised.

>🔹 Generators

Q: What is a generator?
👉 A function that uses yield to return values one at a time. It creates an iterator automatically.

Q: Difference between iterator and generator?
👉

Iterator → needs a class with __iter__ & __next__ methods.

Generator → created using a function with yield (simpler & more memory-efficient).

Q: Why are generators useful?
👉 They allow lazy evaluation (values computed on demand), making them efficient for large datasets.

>🔹 Context Managers

Q: What is a context manager?
👉 A construct that handles setup and teardown of resources using with.

Q: How does it work internally?
👉 Implements __enter__() (setup) and __exit__() (cleanup).

Q: Common use cases?
👉 File handling, database connections, locks, network sessions.

Q: Example of a context manager in Python?
👉 with open("file.txt") as f: ... ensures file is closed automatically.



```python
def same_chars(s1, s2):
    """
    s1 and s2 are strings
    Returns boolean True is a character in s1 is also in s2, and vice
    versa. If a character only exists in one of s1 or s2, returns False.
    """
    # Your code here
    for char in s1:
        if not char in s2:
            return False

    for char in s2:
        if not char in s1:
            return False

    return True

# Examples:
print(same_chars("abc", "cab"))     # prints True
print(same_chars("abccc", "caaab")) # prints True
print(same_chars("abcd", "cabaa"))  # prints False
print(same_chars("abcabc", "cabz")) # prints False
```

```python
def dot_product(tA, tB):
    """
    tA: a tuple of numbers
    tB: a tuple of numbers of the same length as tA
    Assumes tA and tB are the same length.
    Returns a tuple where the:
    * first element is the length of one of the tuples
    * second element is the sum of the pairwise products of tA and tB
    """
    # Your code here
    result = 0
    for i in range(len(tA)):
        result += tA[i] * tB[i]

    return (len(tA), result)

# Examples:
tA = (1, 2, 3)
tB = (4, 5, 6)
print(dot_product(tA, tB)) # prints (3,32)
```

```python
def is_even(n):
    return n % 2 == 0

def divide_5(n):
    return n % 5 == 0

def divide_3(n):
    return n % 3 == 0

def all_true(n, Lf):
    """ n is an int
        Lf is a list of functions that take in an int and return a Boolean
    Returns True if each and every function in Lf returns True when called
    with n as a parameter. Otherwise returns False.
    """
    # Your code here
    for fun in Lf:
        if not fun(n):
            return False
    return True
# Examples:
print(all_true(2, [is_even, divide_3, divide_5]))# prints False
print(all_true(10, [is_even, divide_3, divide_5])) # prints False
print(all_true(30, [is_even, divide_3, divide_5])) # prints True
```

```python
def remove_and_sort(Lin, k):
    """ Lin is a list of ints
        k is an int >= 0
    Mutates Lin to remove the first k elements in Lin and
    then sorts the remaining elements in ascending order.
    If you run out of items to remove, Lin is mutated to an empty list.
    Does not return anything.
    """
    # Your code here
    if k >= len(Lin):
        del Lin[:]
    del Lin[:k]
    Lin.sort()
    return None
# Examples:
L = [1,6,3]
k = 1
remove_and_sort(L, k)
print(L)   # prints the list [3, 6]
```

```python
def count_sqrts(nums_list):
    """
    nums_list: a list
    Assumes that nums_list only contains positive numbers and that there are no duplicates.
    Returns how many elements in nums_list are exact squares of elements in the same list, including itself.
    """
    # Your code here
    result = []

    for num in nums_list:
        if num ** 2 in nums_list:
            result.append(num**2)
    return len(set(result))
# Examples:
print(count_sqrts([3,4,2,1,9,25])) # prints 3
```

```python
def sum_str_lengths(L):
    """
    L is a non-empty list containing either:
    * string elements or
    * a non-empty sublist of string elements
    Returns the sum of the length of all strings in L and
    lengths of strings in the sublists of L. If L contains an
    element that is not a string or a list, or L's sublists
    contain an element that is not a string, raise a ValueError.
    """
    # Your code here
    s = 0

    for item in L:
        if isinstance(item, str):
            s += len(item)
        elif isinstance(item, list):
            s += sum_str_lengths(item)
        else:
            raise ValueError('Check the data type, only string allowed.')
    return s

# Examples:
print(sum_str_lengths(["abcd", "e", "fg"]))  # prints 7
print(sum_str_lengths(["abcd", ["e", "fg"]]))  # prints 7
print(sum_str_lengths([12, ["e", "fg"]]))      # raises ValueError
print(sum_str_lengths(["abcd", [3, "fg"]]))    # raises ValueError
```

```python
import string

# Full character set:
# uppercase + lowercase + digits + punctuation + space
CHARSET = (
    string.ascii_uppercase +
    string.ascii_lowercase +
    string.punctuation +
    " "
)

def caesar_cipher(text: str, shift: int) -> str:
    result = []
    charset_len = len(CHARSET)

    for char in text:
        if char in CHARSET:
            idx = CHARSET.index(char)
            new_idx = (idx + shift) % charset_len
            result.append(CHARSET[new_idx])
        else:
            # If character not in charset, keep it unchanged
            result.append(char)

    return "".join(result)

for i in range(1, 26):
    print(caesar_cipher("Otkz D zvmizy v amzz kjdio! di ocz wjs wzgjr.", i))
```

```python
def flatten(L):
    """
    L: a list
    Returns a copy of L, which is a flattened version of L
    """
    # Your code here
    result = []

    for item in L:
        if isinstance(item, int):
            result.append(item)
        elif isinstance(item, list):
            result += flatten(item)
    return result
# Examples:
L = [[1,4,[6],2],[[[3]],2],4,5]
print(flatten(L)) # prints the list [1,4,6,2,3,2,4,5]
```

```python
def recur_power(base, exp):
    """
    base: int or float.
    exp: int >= 0

    Returns base to the power of exp using recursion.
    Hint: Base case is when exp = 0. Otherwise, in the recursive
    case you return base * base^(exp-1).
    """
    # Your code here
    if exp == 0:
        return 1
    return base * recur_power(base, exp-1)
# Examples:
print(recur_power(2,5))  # prints 32
```

```python
class Circle():
    def __init__(self, radius):
        """ Initializes self with radius """
        # your code here
        self.radius = radius

    def get_radius(self):
        """ Returns the radius of self """
        # your code here
        return self.radius

    def set_radius(self, radius):
        """ radius is a number
        Changes the radius of self to radius """
        # your code here
        self.radius = radius

    def get_area(self):
        """ Returns the area of self using pi = 3.14 """
        # your code here
        pi = 3.14
        return pi * (self.radius) ** 2

    def equal(self, c):
        """ c is a Circle object
        Returns True if self and c have the same radius value """
        # your code here
        return self.radius == c.radius

    def bigger(self, c):
        """ c is a Circle object
        Returns self or c, the Circle object with the bigger radius """
        # your code here
        return self if self.radius > c.radius else c

    def __add__(self, c):
        """ c is a Circle object
        Returns a new Circle object whose radius is
        the sum of self and c's radius """
        # your code here
        return Circle(self.radius + c.radius)

    def __str__(self):
        return f"I am a circle of radius: {self.radius}"
```

```python
c1 = Circle(2)
c2 = Circle(5)
print(c1.bigger(c2))
print(c1+ c2)
```

```python
def all_positive(d):
    """
    d is a dictionary that maps int:list
    Suppose an element in d is a key k mapping to value v (a non-empty list).
    Returns the sorted list of all k whose v elements sums up to a
    positive value.
    """
    # Your code here
    result = []

    for item in d:
        result += [item] if sum(d[item]) > 0 else []
    return result
# Examples:
d = {5:[2,-4], 2:[1,2,3], 1:[2], 4:[1,2,3]}
print(all_positive(d))   # prints the list [1, 2]
```

```python
class Container(object):
    """
    A container object is a list and can store elements of any type
    """
    def __init__(self):
        """
        Initializes an empty list
        """
        self.myList = []

    def size(self):
        """
        Returns the length of the container list
        """
        # Your code here
        return len(self.myList)

    def add(self, elem):
        """
        Adds the elem to one end of the container list, keeping the end
        you add to consistent. Does not return anything
        """
        # Your code here
        self.myList.insert(0, elem)
        return None

    def __str__(self) -> str:
        return f"List with items: {self.myList}"

class Stack(Container):
    """
    A subclass of Container. Has an additional method to remove elements.
    """
    def remove(self):
        """
        The newest element in the container list is removed
        Returns the element removed or None if the queue contains no elements
        """
        # Your code here
        return self.myList.pop(0) if len(self.myList) > 0 else None
```

```python
s = Stack()
s.add(1)
s.add(2)
s.add(3)
print(s)
r = s.remove()
print(r)
print(s)
```
